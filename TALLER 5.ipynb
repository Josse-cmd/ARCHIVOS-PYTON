```python
# EJERCICIO 1 (Comision por ventas + salario al fin de mes)
resp = input("¿Dese saber el valor de su comision? ")
resp = "si"
while resp == "si" or resp == "s" or resp == "yes":
  
  print("Complete los datos")
  salario = float(input("¿Cuanto es su salario?"))
  venta = float(input("¿Cuanto fue el total de la venta?: "))
  comision = venta * 0.10
  total = comision + salario
  print("La comision es= ${:,.0f} y el total de salario a pagar es=${:,.0f}".format(comision, total ))
  resp = input("¿Dese realizar otra consulta? ")
```

```python
#Ejercicio 2 (Descuento 15% por total compra)
resp = input("¿Dese saber su descuento? ")
resp = "si"
while resp == "si" or resp == "s" or resp == "yes":

  venta = float(input("¿Cuanto fue el total de la compra?: "))
  descuento = venta * 0.15
  total = venta - descuento
  print("Su descuento es= ${:,.0f} y el total a pagar es=${:,.0f}".format(descuento, total ))
  resp = input("¿Dese realizar otra consulta? ")
```

```python
# Ejercicio 3 (Calificacion final)
resp = input("¿Dese saber su calificacion final? ")
resp = "si"
while resp == "si" or resp == "s" or resp == "yes":

  calificacion1 = float(input("Digite la calificaicón 1: "))
  calificacion2 = float(input("Digite la calificaicón 2: "))
  calificacion3 = float(input("Digite la calificaicón 3: "))

  prom = round((calificacion1 + calificacion2 + calificacion3) / 3, 2)

  print(f"Su promedio es= {prom}")
  if prom >= 3.5:
    print("Aprueba")
  else:
    print("Reprueba")
  resp = input("¿Dese realizar otra consulta? ")
```

```python
# Ejercicio 4 (Porcentaje de hombres y mujeres en un grupo de estudiantes)

resp = input("¿Dese saber el porcentaje de sus estudiantes por genero? ")
resp = "si"
while resp == "si" or resp == "s" or resp == "yes":
    
  num_est = float(input("Digite el numero de estudiantes: "))
  num_hom = float(input("Digite el numero de hombres: "))
  num_muj = float(input("Digite el numero de mujeres: "))
  total = num_hom + num_muj
  if total <= num_est:
    
    por_muj = (num_muj / num_est) * 100
    por_hom = (num_hom / num_est) * 100

    print("El porcentaje de mujeres es= {:,.0f}% y el porcentaje de hombres es={:,.0f}%".format(por_muj, por_hom ))
    
  else:
    print("Datos incorrectos")
  
  resp = input("¿Dese realizar otra consulta? ")
```
