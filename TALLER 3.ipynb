Ejerciccio 1

```python
#Ejercicio 1 (Si es mayor de 55 años gana bono 5% prepension)

edad = int(input("¿Cual es su edad?: "))

if edad > 55:
  salario = int(input("¿Cual es su salario?: "))
  bono = salario * 0.5
  total = salario + bono
  print("Aplica a bono Prepensional")
  print("Su bono prepensional es de= ${:,.0f} y el total a pagar es=${:,.0f}".format(bono, total ))
else: 
  print("No aplica a bono Prepensional")
```

Ejercicio 3

```python
# Ejercicio 3 (bono segun salario)
salario = int(input("Digite su salario base: "))

if salario >= 1000000 and salario <= 1500000:
  print("Obtiene bono salarial del 2%.")
  bono1 =  salario * 0.02
  total1 = bono1 + salario
  print("Su bono salarial es de= ${:,.0f} y el total a pagar es=${:,.0f}".format(bono1, total1 ))

elif salario >= 1500001 and salario <= 2000000:
  print("Obtiene bono salarial del 5%.")
  bono2 = salario * 0.05
  total2 = bono2 + salario
  print("Su bono salarial es de= ${:,.0f} y el total a pagar es=${:,.0f}".format(bono2, total2 ))

else:
  print("No aplica a bono salarial.")
```
